NUMPY : it uses arrays for its inputs and outputs. It can be extended to objects for matrices. It allows developers to perform fast array processing with minor coding changes

Creating 1D array


```python
import numpy as np
arr = np.array([2,3,4,5,6])
print(arr)
```

2D array

```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
```

Creating Array with Zero

```python
import numpy as np
arr = np.zeros((3, 4))
print(arr)
```

Creating Array with Ones

```python
import numpy as np
arr = np.ones((2,3))
print(arr)
```

Creating a random array

```python
import numpy as np
arr = np.random.rand(2, 3)
print(arr)
```

Accessing element of an array

```python
import numpy as np
arr = np.array([[2,3,4],[4,5,6],[67,8,9]])
print(arr[1, 2])
print("\n")
print(arr[2,0])
```

Slicing an array

```python
import numpy as np
arr = np.array([[2,3,4],[4,5,6],[67,8,9]])
print(arr[1:3])
print("\n")
print(arr[1:])
print("\n")
print(arr[0:2])
print("\n")
print(arr[0:])
```

Adding two array

```python
import numpy as np
a1 = np.array([[1,2],[3,4]])
a2 = np.array([[5,6],[8,9]])
print(a1 + a2)
```

Multiplying two array

```python
import numpy as np
a1 = np.array([[1,2],[3,4]])
a2 = np.array([[5,6],[8,9]])
print(a1 * a2)
```

Finding min and max value of an array

```python
import numpy as np
arr = np.array([2,3,5,7,9])
print(np.min(arr))
print(np.max(arr))
print(np.sum(arr))
print(np.average(arr))
print(np.mean(arr))
```

Pandas Series

Pandas series is a one dimensional labeled array capable of holding data of any type(Integer, string, float, python objects, etc..). Pandas series is nothing but a columns in an excel sheet

Reading A CSV file

```python
#Creating a series from array
#import pandas as pd
import pandas as pd
import numpy as np

#simple array
data = np.array(['a', 'r','r','a','y'])
ser = pd.Series(data)
print(ser)

print('***************************************')

lst = ['a', 'r','r','a','y']
#create series form a list
ser = pd.Series(lst)
print(ser)
```

Pandas DataFreme: is two dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axis(rows and columns). A data frame is two dimentional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of principal components, the data, rows, and columns.


```python
import pandas as pd

lst = ['I', 'Love','Data', 'Science']
#calling Dataframe costructor on list
df = pd.DataFrame(lst)
print(df)
```

2. Visualization Libraries

Matplotlib : It provides an object-oriented API for embedding plot into applications. Easy pyplot function makes some changes to a figure. it creates a figure or plotting area in a figure, plots some lines in a plotting area

```python
#importing matplotlib modele
from matplotlib import pyplot as plt

#x -axis values
x = [5, 2, 9, 4, 7]

#y - axis values
y = [10, 5, 8, 4, 2]

#function to plot
plt.plot(x, y)

#function to show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#creating some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

#Plot the data
plt.plot(x, y)

#Add tittle and axis labels
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('Y')

#Show the plot
plt.show()
```

Seaborn : It is used for making statistical graphics. It provies a high-level interface for drawing attractive and informative graphics. It is very easy to generate in various plots such as heap maps, team series, violin plots.

```python
import seaborn as sns
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x = "total_bill",
            y = "tip",
            col = "time",
            hue = "smoker",
            style = "smoker",
            size = "size",
            data = tips);
```
